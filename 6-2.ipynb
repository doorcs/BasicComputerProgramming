```python
import csv
import datetime

path = '/content/account_book.csv'

while True:
    print("새로 만들기: 1")
    print("가계부 보기: 2")
    print("수입 추가하기: 3")
    print("지출 추가하기: 4")

    cmd = int(input("메뉴를 선택하세요 >> "))

    if cmd == 1:
      print("새로운 가계부를 생성합니다.")
      with open(path, 'w') as file:
        writer = csv.writer(file)
        writer.writerow(['구분', '시간', '금액', '내역', '잔액'])

    elif cmd == 2:
      print("가계부의 내용을 출력합니다.")
      with open(path, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            print(', '.join(row))

    elif cmd == 3:
      print("기존 가계부에 수입을 추가합니다.")
      date = str(datetime.datetime.now())
      amount = int(input("금액을 입력하세요: "))
      info = input("내역을 입력하세요: ")
      current = 0

      with open(path, 'r') as file:
          rows = list(csv.reader(file))
          if len(rows) > 1:
              balance = rows[-1][4]
              current = int(balance)

          current += amount

      records = []
      records.append(["수입", date, amount, info, current])

      with open(path, 'a') as file:
          writer = csv.writer(file)
          writer.writerows(records)

    elif cmd == 4:
      print("기존 가계부에 지출을 추가합니다.")
      date = str(datetime.datetime.now())
      amount = int(input("금액을 입력하세요: "))
      info = input("내역을 입력하세요: ")
      current = 0

      with open(path, 'r') as file:
          rows = list(csv.reader(file))
          if len(rows) > 1:
              balance = rows[-1][4]
              current = int(balance)

          current -= amount

      records = []
      records.append(["지출", date, amount, info, current])

      with open(path, 'a') as file:
          writer = csv.writer(file)
          writer.writerows(records)

    else:  # 교재 예시를 참고하여, 정해진 메뉴 네가지 중 선택하지 않으면 종료되도록 했습니다.
        print("프로그램을 종료합니다.")
        break
```
